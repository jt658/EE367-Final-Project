# Meta-Testing the Network

These were the following 6 meta-test tasks that we ran for our paper. These tasks were performed on synthetic noisy images generated using the test images from the BSDS500 dataset. Most of the noise distributions and levels used during meta-test time were unseen during meta-training. 

```python
noiseParam = ("G", 0.1)
helper.test(numInnerIterations=20, kshot=10, innerLr=10**-3, 
            modelStatePath="best-model2000iter-5layers-withBN-20innerIter.pth.tar", 
            noiseParam= noiseParam,
            folderName="Gaussian_0.1")

noiseParam = ("G", 0.15)
helper.test(numInnerIterations=20, kshot=10, innerLr=10**-3, 
            modelStatePath="best-model2000iter-5layers-withBN-20innerIter.pth.tar", 
            noiseParam= noiseParam,
            folderName="Gaussian_0.15")

noiseParam = ("G", 0.3)
helper.test(numInnerIterations=20, kshot=10, innerLr=10**-3, 
            modelStatePath="best-model2000iter-5layers-withBN-20innerIter.pth.tar", 
            noiseParam= noiseParam,
            folderName="Gaussian_0.3")

noiseParam = ("P", 5)
helper.test(numInnerIterations=20, kshot=10, innerLr=10**-3, 
            modelStatePath="best-model2000iter-5layers-withBN-20innerIter.pth.tar", 
            noiseParam= noiseParam,
            folderName="Poisson_5")

noiseParam = ("P", 7)
helper.test(numInnerIterations=20, kshot=10, innerLr=10**-3, 
            modelStatePath="best-model2000iter-5layers-withBN-20innerIter.pth.tar", 
            noiseParam= noiseParam,
            folderName="Poisson_7")

noiseParam = ("P", 15)
helper.test(numInnerIterations=20, kshot=10, innerLr=10**-3, 
            modelStatePath="best-model2000iter-5layers-withBN-20innerIter.pth.tar", 
            noiseParam= noiseParam,
            folderName="Poisson_15")
```
